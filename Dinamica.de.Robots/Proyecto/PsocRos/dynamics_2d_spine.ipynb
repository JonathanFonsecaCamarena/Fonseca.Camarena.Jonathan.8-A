```python
# Create the image of the simplified 2d spine
# bar length 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
barlength = 0.5 #m

att1 = np.array([0,0])
att2 = np.array([-barlength, barlength])
att3 = np.array([-barlength, -barlength])

CoM = np.array([barlength, 0])
node1 = np.array([barlength * ( 1 - np.cos(np.pi / 3)), barlength * np.sin(np.pi / 2)])
node2 = np.array([barlength * ( 1 - np.cos(np.pi / 3)), -barlength*np.sin(np.pi / 2)])
node3 = np.array([2 * barlength, 0])


#Bars
plt.plot((att1[0],att2[0]),(att1[1],att2[1]), c = 'black', linewidth = 3)
plt.plot((att1[0],att3[0]),(att1[1],att3[1]), c = 'black', linewidth = 3)
plt.plot((CoM[0],node1[0]),(CoM[1],node1[1]), c = 'black', linewidth = 3)
plt.plot((CoM[0],node2[0]),(CoM[1],node2[1]), c = 'black', linewidth = 3)
plt.plot((CoM[0],node3[0]),(CoM[1],node3[1]), c = 'black', linewidth = 3)


# Springs
plt.plot((att1[0],node1[0]),(att1[1],node1[1]), c = 'red', linewidth = 1)
plt.plot((att2[0],node1[0]),(att2[1],node1[1]), c = 'red', linewidth = 1)
plt.plot((att1[0],node2[0]),(att1[1],node2[1]), c = 'red', linewidth = 1)
plt.plot((att3[0],node2[0]),(att3[1],node2[1]), c = 'red', linewidth = 1)

plt.scatter(att1[0], att1[1], c = 'black', linewidth = 6)
plt.scatter(att2[0], att2[1], c = 'black', linewidth = 6)
plt.scatter(att3[0], att3[1], c = 'black', linewidth = 6)
plt.scatter(node1[0], node1[1], c = 'black', linewidth = 6)
plt.scatter(node2[0], node2[1], c = 'black', linewidth = 6)
plt.scatter(node3[0], node3[1], c = 'black', linewidth = 6)
plt.scatter(CoM[0], CoM[1], c = 'black', linewidth = 6)

plt.show()
```
